```python
import pandas as pd
import numpy as np
```

```python
BTS_data = pd.read_csv("billionton_state_download.csv",low_memory = 0)
```

```python
for i in range(0,len(BTS_data.Scenario.unique())):
    print(str(i+1) + ". " + BTS_data.Scenario.unique()[i])
active_case = BTS_data.Scenario.unique()[int(input("     Select active case: "))-1]
```

```python
type_keys = []
for i in range(len(BTS_data)):
    thisEntry = BTS_data.iloc[i]
    type_keys.append(str(thisEntry["Resource Type"]) + "#" + str(thisEntry["Land Source"]) + "#" + str(thisEntry["Feedstock"]))
    if i % 10000 == 0:
        print("    through " + str(i) + " of " + str(len(BTS_data)) + "...")
BTS_data = BTS_data.assign(type_key=type_keys)
BTS_data.iloc[:2]
```
